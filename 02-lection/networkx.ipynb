```python
# import the required pacakges 
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Read the data 
G=nx.read_pajek('dolphins.paj/dolphins.paj'); 

# It seems this graph has multiple edges, let's remove them 
G=nx.Graph(G)
```

# Plot the graph


```python
plt.figure(figsize=(12,8))
pos1=nx.spring_layout(G)
nx.draw(G,pos1,node_color=[0.9,0.4,0.8],edge_color='k',width=3,node_size=500,with_labels=True)
```

# Degree Centrality

```python
deg1d=nx.degree_centrality(G); deg1=np.array(list(deg1d.values()))
plt.figure(figsize=(12,8))
nx.draw(G,pos1,node_color=[0.9,0.4,0.8],edge_color='k',width=2,node_size=15000*deg1,with_labels=True)
```

# Node with maximum degree

```python
import operator
max(deg1d.items(), key=operator.itemgetter(1))[0]
```

# Eigenvector Centrality

```python
eig1d=nx.eigenvector_centrality(G); eig1=np.array(list(eig1d.values()))
plt.figure(figsize=(12,8))
pos=nx.spring_layout(G)
nx.draw(G,pos1,node_color=[0.9,0.4,0.8],edge_color='k',width=2,node_size=15000*eig1,with_labels=True)
```
